```python
# 1. What is hypothesis testing in statistics?

# Ans - Hypothesis testing in statistics is a method used to make decisions or inferences about population parameters based on sample data. It involves 
#       testing an assumption (called a hypothesis) about a population.

#       Hypothesis testing helps you determine whether your sample data provides enough evidence to support a specific claim about a population. It is 
#       widely used in research, quality control, medicine, economics, and more.
```

```python
# 2. What is the null hypothesis, and how does it differ from the alternative hypothesis?

# Ans - The null hypothesis and the alternative hypothesis are the two opposing statements in a hypothesis test. Here's a clear breakdown of their 
#       definitions and differences:

#                   Key Differences

#                      Feature	                   Null Hypothesis (H₀)	                             Alternative Hypothesis (H₁ or Ha)
#                      Meaning	                 No effect or status quo	                        There is an effect or difference
#                   Goal of testing	             To try to disprove it	                            To find evidence supporting it
#                  Relationship to test	         Assumed true unless evidence says otherwise	    Supported if H₀ is rejected
#                    Symbol example	              H₀: μ = 50	                                    H₁: μ ≠ 50 (two-tailed) or μ > 50 (one-tailed)

#      This framework helps ensure decisions are based on statistical evidence rather than assumptions or bias.
```

```python
# 3. What is the significance level in hypothesis testing, and why is it important?

# Ans - The significance level, often denoted by α (alpha), is a threshold used in hypothesis testing to decide whether to reject the null hypothesis 
#       (H₀).The significance level (α) is the probability of making a Type I error, which means rejecting the null hypothesis when it is actually true.

#       It is important because 
#       1. Sets the Standard for Decision Making
#       2. Controls the Risk of False Positives
#       3. Reflects the Tolerance for Error in a Study

#       The significance level is a critical part of hypothesis testing. It defines how strong the evidence must be to reject the null hypothesis and 
#       helps researchers control how often they might be wrong when claiming a significant effect.
```

```python
# 4. What does a P-value represent in hypothesis testing?

# Ans - The p-value in hypothesis testing represents the probability of obtaining a test result at least as extreme as the one actually observed, 
#       assuming the null hypothesis (H₀) is true.

#       The p-value is a crucial part of hypothesis testing—it quantifies the evidence against the null hypothesis. A small p-value suggests that your 
#       observed data is unlikely under H₀, leading you to consider rejecting it.
```

```python
# 5. How do you interpret the P-value in hypothesis testing?

# Ans - The p-value tells you how compatible your data is with the null hypothesis (H₀). It quantifies the probability of observing your results—or 
#       something more extreme—if H₀ is true.

#      Smaller p-value = stronger evidence against H₀.
#      P-value ≠ probability that H₀ is true.
#      Failing to reject H₀ ≠ proving H₀ is true — it just means the evidence isn’t strong enough.

#     The p-value helps you decide whether your sample data provides enough evidence to reject the null hypothesis. A low p-value means your data is 
#     unlikely under H₀, and you may consider the alternative hypothesis instead.
```

```python
# 6. What are Type 1 and Type 2 errors in hypothesis testing?

# Ans - In hypothesis testing, Type I and Type II errors are two kinds of mistakes that can occur when making decisions about the null hypothesis (H₀).

#       Type I Error (False Positive)
#       Definition: Rejecting the null hypothesis (H₀) when it is actually true.
#       You say there is an effect when there isn’t.
#       Symbol: α (alpha) — the significance level
#       Example: A medical test says a patient has a disease (rejects H₀), but they are actually healthy (H₀ was true).
#       Consequence: You conclude something is happening when it's not.

#       Type II Error (False Negative)
#       Definition: Failing to reject the null hypothesis (H₀) when it is actually false.
#       You say there is no effect when there actually is one.
#       Symbol: β (beta)
#       Example: A medical test says a patient is healthy (fail to reject H₀), but they actually have the disease (H₀ was false).
#       Consequence: You miss a real effect or difference.

#       Understanding and balancing these errors is crucial in designing reliable hypothesis tests.
```

```python
# 7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

# Ans - The difference between a one-tailed and a two-tailed test in hypothesis testing lies in the direction of the effect you are testing for.

#       1. One-Tailed Test
#       A one-tailed test checks for an effect in only one direction — either greater than or less than a certain value.
#          Used when you have a specific prediction about the direction of the difference or effect.
#       Example:
#       Suppose a company claims its new battery lasts more than 10 hours.
#       H₀: μ ≤ 10
#       H₁: μ > 10 (Right-tailed test)
#       You're only interested in whether the battery lasts longer, not shorter.
#      Decision Rule:
#      Reject H₀ only if the test statistic falls in the extreme upper (or lower) tail, depending on the direction.

#      2. Two-Tailed Test
#      A two-tailed test checks for an effect in both directions — the value could be greater than or less than a certain point.
#         Used when you do not specify the direction of the effect beforehand.
#      Example:
#      Anyone wants to test if a battery lasts differently from 10 hours (could be more or less).
#      H₀: μ = 10
#      H₁: μ ≠ 10
#      You're interested in any difference from 10 hours, not just in one direction.
#      Decision Rule:
#      Reject H₀ if the test statistic falls in either tail (upper or lower).

#     Use a one-tailed test when you're only interested in an effect in one specific direction.
#     Use a two-tailed test when you're testing for any significant difference, regardless of direction.
#     Choosing the right test before analyzing data is critical to avoid biased results.
```

```python
# 8. What is the Z-test, and when is it used in hypothesis testing?

# Ans - The Z-test is a type of statistical hypothesis test used to determine whether there is a significant difference between sample and population 
#       means, or between the means of two samples, when the population variance is known and the sample size is large.

#       We use the Z-test in cases like When population variance (σ²) is known or When the sample size is large (typically n ≥ 30) or When the data is 
#       approximately normally distributed.

#      The Characteristics of the Z-test are Assuming the data is normally distributed(or approximately normal for large samples due to the Central 
#      Limit Theorem).Population standard deviation (σ) is known and Sample size is typically large (n ≥ 30 is a common rule of thumb).
```

```python
# 9. How do you calculate the Z-score, and what does it represent in hypothesis testing?

# Ans - The Z-score is a standardized value that tells you how many standard deviations a data point (or sample statistic) is from the population mean. 
#       In hypothesis testing, the Z-score quantifies how far a sample statistic (typically a sample mean or proportion) is from the null hypothesis 
#       value, measured in terms of standard errors. It helps determine whether the observed result is statistically significant or likely due to random
#       chance.
```

```python
# 10. What is the T-distribution, and when should it be used instead of the normal distribution?

# Ans - The T-distribution (or Student's t-distribution) is a probability distribution used in statistics when estimating population parameters when the 
#       sample size is small and/or the population standard deviation is unknown.It is similar to the standard normal distribution (Z-distribution) but 
#       has heavier tails, which means it accounts for the extra uncertainty that arises from estimating the standard deviation from the sample data.

#       We can use the T-distribution in hypothesis testing or confidence intervals when:
#       1.The sample size is small (typically n<30).
#       2.The population standard deviation (σ) is unknown.
#       3.The population is assumed to be approximately normally distributed, or the sample is large enough for the Central Limit Theorem to apply.

#       The t-distribution accounts for uncertainty when σ is unknown and sample size is small and It allows more accurate inference under these 
#       conditions.It's essential in small-sample hypothesis testing and confidence interval estimation.
```

```python
# 11. What is the difference between a Z-test and a T-test?

# Ans - The Z-test and T-test are both statistical methods used to test hypotheses about population means, but they differ based on the information
#       available and the sample size.

#       The difference between a Z-test and a T-test are

#           Feature	                             Z-test	                                                 T-test
#          Used when	            Population standard deviation (σ) is known	          Population standard deviation (σ) is unknown
#         Sample size	            Typically large (n ≥ 30)	                          Typically small (n < 30), but can be used for all n
#      Distribution used	        Standard normal distribution (Z-distribution)	      Student’s t-distribution
#    More appropriate when	        You have exact knowledge of population variance       You're using the sample standard deviation (s)
#   Tails of distribution	        Narrower tails	                                      Heavier tails (more conservative with small samples)
#       Converges to Z	            Already normal	                                      Approaches Z-distribution as n → ∞
```

```python
# 14. What is a confidence interval, and how is it used to interpret statistical results?

# Ans - A confidence interval (CI) is a range of values, derived from sample data, that is likely to contain the true population parameter (such as a 
#       mean or proportion) with a certain level of confidence.

#       A confidence interval gives an estimated range that is likely to include the true value of a population parameter. It includes two parts:
#       Point estimate: The best single estimate from the sample (e.g., sample mean).
#       Margin of error: Reflects the uncertainty or variability in the estimate.
#       Confidence Interval = Point Estimate ± Margin of Error

#      A confidence interval(CI) is a tool in statistics that provides a range of plausible values for a population parameter (like a mean or proportion)
#      It allows us to assess both the estimate and the uncertainty around it — making it a key component in interpreting statistical results.
```

```python
# 15. What is the margin of error, and how does it affect the confidence interval?

# Ans - The margin of error (MoE) is a measure of the uncertainty or variability around a sample estimate. It defines how much above and below the point
#       estimate the true population parameter is likely to fall — with a given confidence level.

#       The confidence interval is constructed using the margin of error:
#       Confidence Interval=Point Estimate±Margin of Error
#       1. A larger margin of error → wider confidence interval.
#       2. A smaller margin of error → narrower confidence interval.
#      So, the margin of error directly controls the width of the confidence interval — more error = more uncertainty = wider interval.
```

```python
# 16. How is Bayes' Theorem used in statistics, and what is its significance?

# Ans - Bayes' Theorem is a fundamental rule in probability theory that describes how to update the probability of a hypothesis based on new evidence.
#       Mathematically, it’s expressed as:

#                        P(B∣A)×P(A)
#                P(A∣B)=--------------
#                          P(B)

#       Where:
#       P(A∣B) = Posterior probability: Probability of hypothesis A given data B.
#       P(B∣A) = Likelihood: Probability of observing data B assuming hypothesis A is true.
#       P(A) = Prior probability: Initial probability of hypothesis A before seeing data.
#       P(B) = Marginal likelihood: Total probability of observing data B under all hypotheses.

#       Significance of Bayes' Theorem:

#       Incorporates Prior Knowledge: Allows use of existing knowledge or expert opinion through priors.
#       Dynamic Updating: Continuously updates probabilities as new data arrives — very useful in fields like machine learning, medical diagnosis,
#       and finance.
#       Probabilistic Interpretation: Provides a more intuitive way of reasoning about uncertainty and evidence.
#       Foundation of Bayesian Statistics: Bayes' Theorem is the core of Bayesian methods, an alternative and complement to classical (frequentist) 
#       statistics.
```

```python
# 20. What is an ANOVA test, and what are its assumptions?

# Ans - ANOVA stands for Analysis of Variance. It is a statistical method used to test whether three or more group means are significantly different from
#       each other.Instead of comparing means pairwise (like in multiple t-tests), ANOVA tests all group means simultaneously using variance.

#       Key Assumptions of ANOVA:
#                  Assumption	                                            Explanation
#               1. Independence	                                Observations are independent within and across groups.
#               2. Normality	                                The data in each group are approximately normally distributed.
#               3. Homogeneity of variance (Homoscedasticity)	All groups have equal variances. Tested using Levene’s or Bartlett’s test.

#        ANOVA is fairly robust to violations of normality if sample sizes are equal and large (central limit theorem helps), but homogeneity of variance
#        is more critical, especially when group sizes differ.
```

```python
# 21. What are the different types of ANOVA tests?

# Ans - ANOVA (Analysis of Variance) is a statistical method used to compare the means of two or more groups. There are several types of ANOVA tests, 
#       each suited to different experimental designs and types of data.

#       1. One-Way ANOVA (Single-Factor ANOVA)
#       2. Two-Way ANOVA (Factorial ANOVA)
#       3. Repeated Measures ANOVA
#       4. Mixed-Design ANOVA (Split-Plot ANOVA)
#       5. MANOVA (Multivariate Analysis of Variance)
#       6. ANCOVA (Analysis of Covariance)

#       ANOVA Type	                           Key Feature	                                         Use Case Example
#     One-Way ANOVA	                      One IV with 3+ groups	                       Compare average scores across teaching styles
#     Two-Way ANOVA	                      Two IVs, test interaction	                   Test effect of gender and method on scores
#     Repeated Measures	                  Same subjects, different conditions	       Track blood pressure over 3 time points
#       Mixed ANOVA	                      Combines repeated and independent groups	   Drug vs. placebo measured weekly
#       MANOVA	                          Multiple dependent variables	               Compare test scores in math and reading
#       ANCOVA	                          Controls for covariates	                   Compare scores while adjusting for GPA
```

```python
# 22. What is the F-test, and how does it relate to hypothesis testing?

# Ans - The F-test is a statistical test used to compare two variances or evaluate whether a group of variables significantly affects a dependent 
#       variable. It relies on the F-distribution, which is the ratio of two variances:
#       The F-test is widely used in multiple hypothesis testing scenarios, especially in ANOVA and regression analysis.

#       Role of the F-Test in Hypothesis Testing:
#       In hypothesis testing, you always:
#       1. State a null hypothesis (H₀) and an alternative hypothesis (H₁).
#       2. Calculate a test statistic based on sample data.
#       3. Compare this test statistic to a critical value from a distribution (here, the F-distribution).
#       4. Decide whether to reject or fail to reject H₀.

#       The F-test fits into this framework by producing the F-statistic, which tells you if the variation explained by your model or group differences 
#       is statistically significant.
```

```python
# 2. Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

low = 1     
high = 7     

dist = randint(low, high)

x = np.arange(low, high)
pmf = dist.pmf(x)


plt.figure(figsize=(8, 4))
plt.stem(x, pmf, basefmt=" ", use_line_collection=True)
plt.title('Discrete Uniform Distribution (PMF)')
plt.xlabel('Value')
plt.ylabel('Probability')
plt.xticks(x)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
# 3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

def bernoulli_pdf(x, p):
    """
    Compute the PDF (PMF) of a Bernoulli distribution.

    Parameters:
    x (int): The outcome (0 or 1)
    p (float): Probability of success (0 ≤ p ≤ 1)

    Returns:
    float: Probability of observing outcome x
    """
    if x == 0:
        return 1 - p
    elif x == 1:
        return p
    else:
        return 0.0  
```

```python
# 9. Write a Python function to calculate Z-scores from a dataset and plot them.

import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    """
    Calculate Z-scores from the dataset and plot them.

    Parameters:
    data (list or np.array): Input dataset

    Returns:
    np.array: Array of Z-scores
    """
    data = np.array(data)
    mean = np.mean(data)
    std = np.std(data, ddof=1)  

 
    z_scores = (data - mean) / std

   
    plt.figure(figsize=(8, 4))
    plt.scatter(range(len(data)), z_scores, color='blue')
    plt.axhline(0, color='red', linestyle='--', label='Mean (Z=0)')
    plt.axhline(1, color='green', linestyle='--', label='Z = 1')
    plt.axhline(-1, color='green', linestyle='--', label='Z = -1')
    plt.title('Z-scores of Dataset')
    plt.xlabel('Index')
    plt.ylabel('Z-score')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()

    return z_scores

data = [10, 12, 23, 23, 16, 23, 21, 16]
z_scores = calculate_and_plot_z_scores(data)
print("Z-scores:", z_scores)
```

```python
# 10. Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set seaborn style for nicer plots
sns.set(style="whitegrid")

def demonstrate_clt(sample_size=30, num_samples=1000):
    """
    Demonstrate the Central Limit Theorem with a non-normal distribution.

    Parameters:
    sample_size (int): Number of observations per sample
    num_samples (int): Number of samples to draw
    """
  
 
    data = np.random.exponential(scale=1.0, size=(num_samples, sample_size))

 
    sample_means = data.mean(axis=1)

 
    plt.figure(figsize=(14, 6))

    plt.subplot(1, 2, 1)
    sns.histplot(data.flatten(), bins=50, kde=True, color='orange')
    plt.title('Original Data Distribution (Exponential)')
    plt.xlabel('Value')
    plt.ylabel('Frequency')


    plt.subplot(1, 2, 2)
    sns.histplot(sample_means, bins=30, kde=True, color='blue')
    plt.title(f'Distribution of Sample Means (n={sample_size})')
    plt.xlabel('Sample Mean')
    plt.ylabel('Frequency')

    plt.tight_layout()
    plt.show()


demonstrate_clt(sample_size=30, num_samples=1000)
```

```python
# 15. Implement hypothesis testing using Z-statistics for a sample dataset.

import numpy as np
from scipy.stats import norm

def z_test(sample_data, population_mean, population_std, alpha=0.05):
    """
    Perform a one-sample Z-test.
    
    Parameters:
    sample_data (list or np.array): The sample observations
    population_mean (float): The hypothesized population mean (H0)
    population_std (float): The known population standard deviation
    alpha (float): Significance level (default = 0.05)
    
    Returns:
    None
    """
    sample_data = np.array(sample_data)
    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)

    
    z = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    
    z_critical = norm.ppf(1 - alpha/2)

    print(f"Sample Mean = {sample_mean:.2f}")
    print(f"Z-statistic = {z:.4f}")
    print(f"Critical Z for alpha={alpha} (two-tailed) = ±{z_critical:.4f}")

   
    if abs(z) > z_critical:
        print("Reject the null hypothesis (H0): Significant difference.")
    else:
        print("Fail to reject the null hypothesis (H0): No significant difference.")


sample = [72, 69, 75, 71, 70, 74, 68, 73, 70, 71]
pop_mean = 70      
pop_std = 3        
z_test(sample, pop_mean, pop_std)
```

```python
# 16. Create a confidence interval for a dataset using Python and interpret the result.

import numpy as np
from scipy import stats

def confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for the mean of a dataset using the t-distribution.

    Parameters:
    data (list or np.array): Sample dataset
    confidence (float): Confidence level (default 0.95 for 95%)

    Returns:
    tuple: (lower_bound, upper_bound)
    """
    data = np.array(data)
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)  # Standard error of the mean
    t_critical = stats.t.ppf((1 + confidence) / 2, df=n-1)

    margin_of_error = t_critical * std_err
    lower = mean - margin_of_error
    upper = mean + margin_of_error

    print(f"Sample Mean = {mean:.2f}")
    print(f"{int(confidence*100)}% Confidence Interval: ({lower:.2f}, {upper:.2f})")
    
    return (lower, upper)

sample_data = [72, 69, 75, 71, 70, 74, 68, 73, 70, 71]
confidence_interval(sample_data, confidence=0.95)
```

```python
# 1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.

import numpy as np
from scipy.stats import norm

def z_test(sample_data, pop_mean, pop_std, alpha=0.05):
    """
    Perform a Z-test for comparing sample mean to a known population mean.

    Parameters:
    - sample_data: array-like, sample observations
    - pop_mean: known population mean
    - pop_std: known population standard deviation
    - alpha: significance level (default 0.05)

    Returns:
    - z_stat: computed Z statistic
    - p_value: two-tailed p-value
    """
    n = len(sample_data)
    sample_mean = np.mean(sample_data)
    std_error = pop_std / np.sqrt(n)

    
    z_stat = (sample_mean - pop_mean) / std_error


    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    print(f"Sample Mean = {sample_mean:.4f}")
    print(f"Population Mean = {pop_mean:.4f}")
    print(f"Z-statistic = {z_stat:.4f}")
    print(f"P-value = {p_value:.4f}")

    if p_value < alpha:
        print(f"Reject the null hypothesis at α = {alpha}")
        print("There is sufficient evidence that the sample mean is different from the population mean.")
    else:
        print(f"Fail to reject the null hypothesis at α = {alpha}")
        print("There is insufficient evidence to conclude the sample mean differs from the population mean.")

    return z_stat, p_value

np.random.seed(0)

sample = np.random.normal(loc=105, scale=15, size=50)

population_mean = 100
population_std = 15

z_test(sample, population_mean, population_std)
```

```python
# 2. Simulate random data to perform hypothesis testing and calculate the corresponding P-value using Python.

import numpy as np
from scipy.stats import ttest_ind

def simulate_and_test(sample_size=50, mean1=100, mean2=105, std=15, alpha=0.05):
    """
    Simulate two independent samples and perform a two-sample t-test.

    Parameters:
    - sample_size: number of observations in each sample
    - mean1: mean of first population
    - mean2: mean of second population
    - std: standard deviation of both populations (assumed equal)
    - alpha: significance level

    Returns:
    - p_value: p-value from the t-test
    """
    np.random.seed(42)
 
    sample1 = np.random.normal(loc=mean1, scale=std, size=sample_size)
    sample2 = np.random.normal(loc=mean2, scale=std, size=sample_size)
   
    t_stat, p_value = ttest_ind(sample1, sample2, equal_var=True)

    print(f"Sample 1 Mean: {np.mean(sample1):.2f}")
    print(f"Sample 2 Mean: {np.mean(sample2):.2f}")
    print(f"T-statistic: {t_stat:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print(f"Reject the null hypothesis at α = {alpha}.")
        print("There is a statistically significant difference between the two means.")
    else:
        print(f"Fail to reject the null hypothesis at α = {alpha}.")
        print("There is no statistically significant difference between the two means.")

    return p_value

simulate_and_test()
```

```python
# 5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_type1_type2_errors(mu0, mu1, sigma, n, alpha=0.05, tail='two-sided'):
    """
    Plots Type I and Type II error regions for hypothesis testing.

    Parameters:
    - mu0: Null hypothesis mean
    - mu1: Alternative hypothesis mean
    - sigma: Population standard deviation
    - n: Sample size
    - alpha: Significance level (default 0.05)
    - tail: 'two-sided', 'left', or 'right'
    """
    se = sigma / np.sqrt(n)
    

    if tail == 'two-sided':
        z_crit = norm.ppf(1 - alpha / 2)
        lower_crit = mu0 - z_crit * se
        upper_crit = mu0 + z_crit * se
    elif tail == 'left':
        z_crit = norm.ppf(alpha)
        lower_crit = mu0 + z_crit * se
        upper_crit = None
    elif tail == 'right':
        z_crit = norm.ppf(1 - alpha)
        upper_crit = mu0 + z_crit * se
        lower_crit = None
    else:
        raise ValueError("tail must be 'two-sided', 'left', or 'right'")

 
    x = np.linspace(mu0 - 4*se, mu1 + 4*se, 1000)
    null_dist = norm.pdf(x, mu0, se)
    alt_dist = norm.pdf(x, mu1, se)


    plt.figure(figsize=(10, 6))
    plt.plot(x, null_dist, label="Null Hypothesis", color='black')
    plt.plot(x, alt_dist, label="Alternative Hypothesis", color='blue')

  
    if tail == 'two-sided':
        plt.fill_between(x, 0, null_dist, where=(x < lower_crit) | (x > upper_crit), 
                         color='red', alpha=0.3, label='Type I Error (α)')
    elif tail == 'left':
        plt.fill_between(x, 0, null_dist, where=(x < lower_crit), 
                         color='red', alpha=0.3, label='Type I Error (α)')
    elif tail == 'right':
        plt.fill_between(x, 0, null_dist, where=(x > upper_crit), 
                         color='red', alpha=0.3, label='Type I Error (α)')


    if tail == 'two-sided':
        plt.fill_between(x, 0, alt_dist, where=(x >= lower_crit) & (x <= upper_crit), 
                         color='blue', alpha=0.3, label='Type II Error (β)')
    elif tail == 'left':
        plt.fill_between(x, 0, alt_dist, where=(x >= lower_crit), 
                         color='blue', alpha=0.3, label='Type II Error (β)')
    elif tail == 'right':
        plt.fill_between(x, 0, alt_dist, where=(x <= upper_crit), 
                         color='blue', alpha=0.3, label='Type II Error (β)')

    plt.axvline(mu0, color='black', linestyle='--', label='μ₀')
    plt.axvline(mu1, color='blue', linestyle='--', label='μ₁')

    plt.title("Type I and Type II Errors in Hypothesis Testing")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()

plot_type1_type2_errors(mu0=100, mu1=105, sigma=15, n=36, alpha=0.05, tail='two-sided')
```

```python
# 6. Write a Python program to perform an independent T-test and interpret the results.

import numpy as np
from scipy.stats import ttest_ind

def perform_independent_ttest(group1, group2, alpha=0.05, equal_var=True):
    """
    Performs an independent t-test between two groups and interprets the result.

    Parameters:
    - group1: list or array of sample values (group 1)
    - group2: list or array of sample values (group 2)
    - alpha: significance level (default 0.05)
    - equal_var: assume equal variances (default True)
    """

   
    t_stat, p_value = ttest_ind(group1, group2, equal_var=equal_var)

   
    print(f"Group 1 mean: {np.mean(group1):.2f}")
    print(f"Group 2 mean: {np.mean(group2):.2f}")
    print(f"\nT-statistic: {t_stat:.4f}")
    print(f"P-value: {p_value:.4f}")
    
  
    print(f"\nSignificance level: α = {alpha}")
    if p_value < alpha:
        print("🟢 Result: Reject the null hypothesis (statistically significant difference).")
    else:
        print("🔵 Result: Fail to reject the null hypothesis (no statistically significant difference).")


group_a = [20, 22, 19, 24, 25, 23, 20]
group_b = [28, 27, 30, 29, 31, 33, 35]

perform_independent_ttest(group_a, group_b, alpha=0.05, equal_var=True)
```

```python
# 8. Simulate data and perform both Z-test and T-test, then compare the results using Python.

import numpy as np
from scipy.stats import ttest_1samp, norm
import matplotlib.pyplot as plt

def simulate_and_compare_tests(mu_pop=100, sigma_pop=15, n=30, alpha=0.05):
    
    np.random.seed(42)
    sample = np.random.normal(loc=103, scale=15, size=n)  # sample with true mean ≠ mu_pop
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)

    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Sample Std Dev: {sample_std:.2f}")


    se_z = sigma_pop / np.sqrt(n)
    z_stat = (sample_mean - mu_pop) / se_z
    p_value_z = 2 * (1 - norm.cdf(abs(z_stat)))

   
    t_stat, p_value_t = ttest_1samp(sample, mu_pop)

  
    print("\n--- Z-Test (σ known) ---")
    print(f"Z-statistic: {z_stat:.4f}")
    print(f"P-value: {p_value_z:.4f}")
    if p_value_z < alpha:
        print("🟢 Reject H₀ (Z-test)")
    else:
        print("🔵 Fail to reject H₀ (Z-test)")

    print("\n--- T-Test (σ unknown) ---")
    print(f"T-statistic: {t_stat:.4f}")
    print(f"P-value: {p_value_t:.4f}")
    if p_value_t < alpha:
        print("🟢 Reject H₀ (T-test)")
    else:
        print("🔵 Fail to reject H₀ (T-test)")

   
    labels = ['Z-test', 'T-test']
    stats = [abs(z_stat), abs(t_stat)]
    p_values = [p_value_z, p_value_t]

    plt.figure(figsize=(8, 4))
    plt.bar(labels, stats, color=['orange', 'blue'], alpha=0.7)
    plt.axhline(y=norm.ppf(1 - alpha/2), color='red', linestyle='--', label='Critical value (Z, α=0.05)')
    plt.title('Comparison of Z-test and T-test Statistics')
    plt.ylabel('Test Statistic (Absolute Value)')
    plt.legend()
    plt.tight_layout()
    plt.show()

simulate_and_compare_tests()
```

```python
# 9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

import numpy as np
from scipy.stats import t

def confidence_interval(data, confidence=0.95):
    """
    Calculates the confidence interval for the mean of a sample.

    Parameters:
    - data: array-like, sample data
    - confidence: confidence level (default is 0.95)

    Returns:
    - (lower_bound, upper_bound): tuple representing the confidence interval
    """
    data = np.array(data)
    n = len(data)
    mean = np.mean(data)
    sem = np.std(data, ddof=1) / np.sqrt(n)  
    df = n - 1
    t_crit = t.ppf((1 + confidence) / 2, df)  

    margin_of_error = t_crit * sem
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    return lower_bound, upper_bound

sample_data = [105, 98, 102, 100, 95, 108, 101, 99]
ci = confidence_interval(sample_data, confidence=0.95)
print(f"95% Confidence Interval for the mean: {ci}")
```

```python
# 10. Write a Python program to calculate the margin of error for a given confidence level using sample data.

import numpy as np
from scipy.stats import t

def calculate_margin_of_error(data, confidence=0.95):
    """
    Calculates the margin of error for a given confidence level using sample data.

    Parameters:
    - data: list or array of sample values
    - confidence: confidence level (default is 0.95)

    Returns:
    - margin_of_error: the margin of error value
    """
    data = np.array(data)
    n = len(data)
    mean = np.mean(data)
    std_dev = np.std(data, ddof=1)
    sem = std_dev / np.sqrt(n)
    
    df = n - 1  
    t_crit = t.ppf((1 + confidence) / 2, df)

    moe = t_crit * sem

    print(f"Sample size (n): {n}")
    print(f"Sample mean: {mean:.2f}")
    print(f"Sample standard deviation: {std_dev:.2f}")
    print(f"Standard error: {sem:.2f}")
    print(f"T critical value (t*): {t_crit:.4f}")
    print(f"Confidence Level: {int(confidence * 100)}%")
    print(f"✅ Margin of Error: ±{moe:.2f}")

    return moe

sample_data = [102, 98, 105, 100, 97, 103, 101, 99]
calculate_margin_of_error(sample_data, confidence=0.95)
```

```python
# 11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

def bayesian_inference(prior, likelihood, false_positive_rate):
    """
    Computes posterior probability using Bayes' Theorem.
    
    Parameters:
    - prior: P(H), prior probability of having the disease
    - likelihood: P(E|H), probability of testing positive if diseased
    - false_positive_rate: P(E|~H), probability of testing positive if NOT diseased
    
    Returns:
    - posterior: P(H|E), probability of disease given a positive test
    """
  
    prior_complement = 1 - prior

    
    marginal = likelihood * prior + false_positive_rate * prior_complement

    
    posterior = (likelihood * prior) / marginal

    return posterior


P_D = 0.01               
P_Pos_given_D = 0.99     
P_Pos_given_no_D = 0.05  

posterior = bayesian_inference(P_D, P_Pos_given_D, P_Pos_given_no_D)

print(f"Posterior Probability (P(Disease | Positive)) = {posterior:.4f} or {posterior * 100:.2f}%")
```

```python
# 12. Perform a Chi-square test for independence between two categorical variables in Python.

import pandas as pd
import scipy.stats as stats



data = {
    'Tea':    [30, 20],  
    'Coffee': [20, 30]
}
table = pd.DataFrame(data, index=['Male', 'Female'])

print("Contingency Table:")
print(table)


chi2, p, dof, expected = stats.chi2_contingency(table)


print("\n--- Chi-Square Test Result ---")
print(f"Chi2 Statistic = {chi2:.4f}")
print(f"Degrees of Freedom = {dof}")
print(f"P-value = {p:.4f}")
print("\nExpected Frequencies:")
print(pd.DataFrame(expected, index=table.index, columns=table.columns))


alpha = 0.05
if p < alpha:
    print("\n🟢 Reject the null hypothesis: There IS a relationship between Gender and Beverage Preference.")
else:
    print("\n🔵 Fail to reject the null hypothesis: No significant relationship between Gender and Beverage Preference.")
```

```python
# 13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data.

import numpy as np
import pandas as pd

def calculate_expected_frequencies(observed):
    """
    Calculate expected frequencies for a Chi-square test given observed frequencies.

    Parameters:
    - observed: 2D array-like or pandas DataFrame of observed counts

    Returns:
    - expected: pandas DataFrame of expected frequencies
    """
    observed = np.array(observed)
    row_totals = observed.sum(axis=1).reshape(-1, 1)  
    col_totals = observed.sum(axis=0).reshape(1, -1)  
    grand_total = observed.sum()

    expected = (row_totals @ col_totals) / grand_total 

 
    if isinstance(observed, pd.DataFrame):
        expected = pd.DataFrame(expected, index=observed.index, columns=observed.columns)
    return expected


observed = pd.DataFrame({
    'Tea': [30, 20],
    'Coffee': [20, 30]
}, index=['Male', 'Female'])

expected = calculate_expected_frequencies(observed)
print("Observed Frequencies:")
print(observed)
print("\nExpected Frequencies:")
print(expected)
```
